```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Ronel2030/Ronel/main/Gold%20Price%20Prediction.csv")
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df['DXY']=pd.to_datetime(df['DXY'],errors = 'coerce')
```

```python
df['DXY'].info()
```

```python
df['Crude']=pd.to_datetime(df['Crude'],errors = 'coerce')
```

```python
df['Crude'].info()
```

```python
df['date_gap'] = df['DXY'] - df['Crude']
```

```python
df['Date'] = pd.to_numeric(df['Date'],errors= 'coerce').astype('Int64')
```
